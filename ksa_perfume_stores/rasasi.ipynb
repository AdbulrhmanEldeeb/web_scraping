```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_rasasi_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
rasasi_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
rasasi_stores_df['rating']=rasasi_stores_df['rating'].str.replace('نجمة','')
# rasasi_stores_df=rasasi_stores_df[rasasi_stores_df['store_name'].str.contains('أجمل|اجمل|rasasi|rasasi ',regex=True)]
# rasasi_stores_df=rasasi_stores_df[~rasasi_stores_df['store_name'].str.contains('صالون|للمفروشات|فراح|مناسبات|بوفيه|أنتي',regex=True)]
rasasi_stores_df=rasasi_stores_df[rasasi_stores_df['location'].str.contains('المملكة العربية السعودية')]
rasasi_stores_df['rating']=pd.to_numeric(rasasi_stores_df['rating'])
rasasi_stores_df
```
