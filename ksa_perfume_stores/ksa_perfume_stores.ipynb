```python
all_alarabya_stores=[*alarabya_list,*alarabya_list_1,*alarabya_list_2,*alarabya_list_3,*alarabya_list_4,*alarabya_list_5,\
                     *alarabya_list_7,*alarabya_list_8,*alarabya_list_9,*alarabya_list_10,*alarabya_list_11,*alarabya_list_12,*alarabya_list_13,\
                     *alarabya_list_14,*alarabya_list_15,*alarabya_list_16,*alarabya_list_17,*alarabya_list_18,*alarabya_list_19
                    ]
print(len(all_alarabya_stores))
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_alarabya_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
len(stores_names),len(locations),len(ratings)
```

```python
import pandas as pd 
alarabia_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
# تنظيف البيانات قبل دمجها
alarabia_stores_df['store_name'] = alarabia_stores_df['store_name'].str.strip().str.lower()
alarabia_stores_df['location'] = alarabia_stores_df['location'].str.strip().str.lower()
alarabia_stores_df['rating']=alarabia_stores_df['rating'].apply(lambda x : x.replace('نجمة','').strip())
alarabia_stores_df['rating']=alarabia_stores_df['rating'].astype('float16')
alarabia_stores_df['rating']=alarabia_stores_df['rating'].apply(lambda x : round(x,1))
import re
import warnings 
warnings.filterwarnings('ignore')
# List of unwanted store names
unwanted_names = ['الدخيل للعود', 'الماجد للعود', 'درعه DERAAH', 'السامر للعود', 
                  'محمود للعود', 'Abdul Samad Al Qurashi', 'Dkhoun | دخون', 
                  'OudElite - نخبة العود 7345', 'عطور رسيس | Rasees perfumes', 
                  'Almajed For Oud', 'نخبة العود','دخون']

# Join unwanted names into a single regex pattern
unwanted_pattern = '|'.join(map(re.escape, unwanted_names))

# Iterate over the DataFrame and drop rows where the store name matches any unwanted name
for index, row in alarabia_stores_df.iterrows():
    store_name = row['store_name']  # Replace with your actual column name for store names
    if re.search(unwanted_pattern, store_name):
        alarabia_stores_df.drop(index, inplace=True)

# Alternatively, you can achieve the same in a vectorized way:
alarabia_stores_df = alarabia_stores_df[~alarabia_stores_df['store_name'].str.contains(unwanted_pattern, regex=True)]

alarabia_stores_df
```
