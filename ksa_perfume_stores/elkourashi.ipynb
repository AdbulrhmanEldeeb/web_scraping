```python
all_elkourashi_stores=[*elkourashi_list,*elkourashi_list_1,*elkourashi_list_2,*elkourashi_list_3,*elkourashi_list_4,
                      *elkourashi_list_5,*elkourashi_list_5,*elkourashi_list_6,*elkourashi_list_7,*elkourashi_list_8,*elkourashi_list_9,
                      *elkourashi_list_10,*elkourashi_list_11,*elkourashi_list_12,*elkourashi_list_13,*elkourashi_list_14,*elkourashi_list_15]
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_elkourashi_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
elkourashi_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
elkourashi_stores_df['rating']=elkourashi_stores_df['rating'].str.replace('نجمة','')
elkourashi_stores_df=elkourashi_stores_df[elkourashi_stores_df['store_name'].str.contains('عبدالصمد|Abdul Samad|عبد الصمد|القرشي|Qurashi',regex=True)]
elkourashi_stores_df=elkourashi_stores_df[~elkourashi_stores_df['location'].str.contains('عبدالصمد|Abdul Samad|عبد الصمد|القرشي|Qurashi',regex=True)]
elkourashi_stores_df=elkourashi_stores_df[elkourashi_stores_df['location'].str.contains('المملكة العربية السعودية')]
elkourashi_stores_df['rating']=pd.to_numeric(elkourashi_stores_df['rating'])
elkourashi_stores_df
```
