```python
all_almajed_stores=[*almajed_list,*almajed_list_1,*almajed_list_2,*almajed_list_3,*almajed_list_4,
                      *almajed_list_5,*almajed_list_5,*almajed_list_6,*almajed_list_7,*almajed_list_8,*almajed_list_9,
                      *almajed_list_10]
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_almajed_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
almajed_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
almajed_stores_df['rating']=almajed_stores_df['rating'].str.replace('نجمة','')
almajed_stores_df=almajed_stores_df[almajed_stores_df['store_name'].str.contains('الماجد للعود|Almajed|ماجد|الماجد',regex=True)]
almajed_stores_df=almajed_stores_df[almajed_stores_df['location'].str.contains('المملكة العربية السعودية')]
almajed_stores_df['rating']=pd.to_numeric(almajed_stores_df['rating'])
almajed_stores_df
```
