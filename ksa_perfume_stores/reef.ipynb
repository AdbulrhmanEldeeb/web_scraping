```python
all_reef_stores=[*reef_list,*reef_list_1,*reef_list_3,*reef_list_4,
                      *reef_list_5,*reef_list_5,*reef_list_6,*reef_list_7,*reef_list_8,*reef_list_9]
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_reef_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
reef_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
reef_stores_df['rating']=reef_stores_df['rating'].str.replace('نجمة','')
# reef_stores_df=reef_stores_df[reef_stores_df['store_name'].str.contains('ريف|اجمل|reef| ',regex=True)]
# reef_stores_df=reef_stores_df[~reef_stores_df['store_name'].str.contains('صالون|للمفروشات|فراح|مناسبات|بوفيه|أنتي',regex=True)]
reef_stores_df=reef_stores_df[reef_stores_df['location'].str.contains('المملكة العربية السعودية')]
reef_stores_df['rating']=pd.to_numeric(reef_stores_df['rating'])
reef_stores_df
```
