```python
all_ebrahim_stores=[*ebrahim_list,*ebrahim_list_1,*ebrahim_list_3,*ebrahim_list_4,
                      *ebrahim_list_5,*ebrahim_list_5,*ebrahim_list_6,*ebrahim_list_7,*ebrahim_list_8,]
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_ebrahim_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
ebrahim_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
ebrahim_stores_df['rating']=ebrahim_stores_df['rating'].str.replace('نجمة','')
# ebrahim_stores_df=ebrahim_stores_df[ebrahim_stores_df['store_name'].str.contains('ريف|اجمل|ebrahim| ',regex=True)]
ebrahim_stores_df=ebrahim_stores_df[~ebrahim_stores_df['store_name'].str.contains('عبد الصمد|عبدالصمد ',regex=True)]
ebrahim_stores_df=ebrahim_stores_df[ebrahim_stores_df['location'].str.contains('المملكة العربية السعودية')]
ebrahim_stores_df['rating']=pd.to_numeric(ebrahim_stores_df['rating'])
ebrahim_stores_df
```
