```python
all_derah_stores=[*derah_list,*derah_list_1,*derah_list_2,*derah_list_3,*derah_list_4,*derah_list_5,*derah_list_6 ,*derah_list_7,*derah_list_8,
                  *derah_list_9,*derah_list_10,*derah_list_11,*derah_list_12,*derah_list_13,*derah_list_14,*derah_list_15,*derah_list_16,*derah_list_17,
                    *derah_list_18]
print(len(all_derah_stores))
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_derah_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
derah_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
derah_stores_df['rating']=derah_stores_df['rating'].str.replace('نجمة','')
derah_stores_df=derah_stores_df[derah_stores_df['store_name'].str.contains('درعه|DERAAH|deraah',regex=True)]
derah_stores_df['rating']=pd.to_numeric(derah_stores_df['rating'])
derah_stores_df
```
