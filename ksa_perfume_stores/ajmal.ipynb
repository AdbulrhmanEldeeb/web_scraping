```python
all_ajmal_stores=[*ajmal_list,*ajmal_list_1,*ajmal_list_2,*ajmal_list_3,*ajmal_list_4,
                      *ajmal_list_5,*ajmal_list_5,*ajmal_list_6,*ajmal_list_7,*ajmal_list_8,*ajmal_list_9]
```

```python
stores_names=[]
locations=[]
ratings=[]
for dictionary in all_ajmal_stores:
        stores_names.append(dictionary['store_name']) if dictionary['store_name'] else ' '
        locations.append(dictionary['location']) if dictionary['location'] else ' '
        ratings.append(dictionary['rating']) if dictionary['rating'] else ' '
```

```python
import pandas as pd 
ajmal_stores_df=pd.DataFrame({'store_name':stores_names,'location':locations,'rating':ratings})
ajmal_stores_df['rating']=ajmal_stores_df['rating'].str.replace('نجمة','')
ajmal_stores_df=ajmal_stores_df[ajmal_stores_df['store_name'].str.contains('أجمل|اجمل|Ajmal|ajmal ',regex=True)]
ajmal_stores_df=ajmal_stores_df[~ajmal_stores_df['store_name'].str.contains('صالون|للمفروشات|فراح|مناسبات|بوفيه|أنتي',regex=True)]
ajmal_stores_df=ajmal_stores_df[ajmal_stores_df['location'].str.contains('المملكة العربية السعودية')]
ajmal_stores_df['rating']=pd.to_numeric(ajmal_stores_df['rating'])
ajmal_stores_df
```
